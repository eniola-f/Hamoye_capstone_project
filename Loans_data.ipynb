```python
# Bankruptcies with NaN have 0.0 tax liens
data[data["Bankruptcies"].isna()].sample(10)
```

```python
# Bankruptcies with 0 values also have 0.0 tax liens
data[data["Bankruptcies"] == 0].sample(10)
```

```python
# Filling in missing Bankruptcies values with 0.0 since tax liens is 0.0
data["Bankruptcies"].fillna(0, inplace=True)
```

```python
# Bankruptcies columns have no more missing values
data["Bankruptcies"].isna().sum()
```

```python
# Sampling all the rows with missing tax liens
data[data["Tax Liens"].isna()].sample(9)
```

```python
# Filling in the missing tax liens with 0
data["Tax Liens"].fillna(0, inplace=True)
```

```python
data.isna().sum()
```

```python
# How to calculate credit score. Do we calculate the credit scores or fill in the mean?
```

```python
# Fillin in the mean credit scores
data["Credit Score"].fillna(np.mean(data["Credit Score"]), inplace=True)
```

```python
data["Credit Score"].isna().sum()
```

```python
# Loans for customers with credit score above 5000 were Charged Off
data["Loan Status"][data["Credit Score"] > 5000].value_counts()
```

```python
# Loans for customers with credit score below 
data["Loan Status"][data["Credit Score"] <5000].value_counts()
```

```python
# All other credit scores are below 1000
data["Loan Status"][data["Credit Score"] <5000].value_counts() == data["Loan Status"][data["Credit Score"] <1000].value_counts()
```

```python
# Comparing the target values; Charged off and fully paid. Data is imbalanced. We can downsample
sns.countplot(data=data, x=data["Loan Status"])
```
